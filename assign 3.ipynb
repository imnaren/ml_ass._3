# Aspiring mind Employability outcome 2015

dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.

## 3. Feature transformation.
Like standardization of numerical columns and dummy variable or encoding for categorical columns.

```python
# First remove unneccesary columns
ns.drop(['Unnamed: 0', 'ID', 'CollegeID', 'CollegeCityID'], axis=1, inplace=True)
ns.columns
```

```python
# first- ecode binary categorical columns.

binary_category = ns[['Gender', 'CollegeTier', 'CollegeCityTier']]
ns.drop(['Gender', 'CollegeTier', 'CollegeCityTier'], axis=1, inplace=True)          # remove these columns from main

pd.set_option('mode.chained_assignment', None)
binary_category.loc[:, 'Gender'] = binary_category['Gender'].map({'m':1, 'f':0})     # map 'm' to 1 and 'f' to 0
binary_category.loc[:, 'CollegeTier'] = binary_category['CollegeTier'] - 1       # also converting college tier to binary

binary_category.head()
```

```python
# Second- encode categorical columns other than binary.
other_category = ns[['Designation', 'JobCity', 'Degree', 'Specialization', 'CollegeState', '12board', '10board']]
ns.drop(['Designation', 'JobCity', 'Degree', 'Specialization', 'CollegeState', '12board', '10board'], axis=1, inplace=True)

# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

# column names are (annoyingly) lost after OneHotEncoding
# (i.e. the dataframe is converted to a numpy ndarray)

other_category = pd.DataFrame(encoder.fit_transform(other_category), 
                      columns=encoder.get_feature_names(other_category.columns))

other_category.head()
```

```python
# Third: Now perform standardization on numerical columns.

numerical_data = ns.select_dtypes(include=['int64', 'float64'])  # create dataframe with numerical columns.
ns.drop(numerical_data.columns, axis=1, inplace=True)            # remove these column from main dataframe

# import StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# standardization and datafram creation
numerical_data = pd.DataFrame(scaler.fit_transform(numerical_data), columns=numerical_data.columns)

numerical_data.head()
```

now,
    1. df contains only date columns.
    2. binary_category contains only binary encoded columns.
    3. other_category contains all other categorical column on with OneHotEncoding is applied.
    4. numerical_data contains standardized numerical columns.
Let's merge all the dataframe back to get the final data.

```python
final_ns = pd.concat([ns, binary_category, other_category, numerical_data], axis=1)
print("Rows: ", final_ns.shape[0])
print("Columns: ", final_ns.shape[1])
final_ns.head()
```
